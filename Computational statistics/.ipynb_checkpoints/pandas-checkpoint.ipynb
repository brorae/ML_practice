```python
import numpy as np
import pandas as pd
from scipy import stats
from matplotlib import pyplot as plt
```

```python
sample = pd.read_excel("./sample.xlsx")
sample.head()
```

```python
sampleA = sample['Medicine'] == 'A'
sampleB = sample['Medicine'] == 'B'
sampleC = sample['Medicine'] == 'C'
sampleA = sample[sampleA]
sampleB = sample[sampleB]
sampleC = sample[sampleC]
sampleA.head()
sampleB.head()
sampleC.head()
```

```python
import operator
sampleA_Effect = sampleA["Effect"].mean()
sampleA_Problem = sampleA["Problem"].mean()
sampleB_Effect = sampleB["Effect"].mean()
sampleB_Problem = sampleB["Problem"].mean()
sampleC_Effect = sampleC["Effect"].mean()
sampleC_Problem = sampleC["Problem"].mean()
sample_mean_effect = {'A':sampleA_Effect,'B':sampleB_Effect,'C':sampleC_Effect}
sample_mean_problem = {'A':sampleA_Problem,'B':sampleB_Problem,'C':sampleC_Problem}
sample_mean_problem = sorted(sample_mean_problem.items(),key=operator.itemgetter(1),reverse=True)

print(sample_mean_effect)
for i in range(3):
    print(sample_mean_problem[i][0])
```
