```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.datasets import make_circles
```

```python
X,y = make_circles(n_samples=500,factor=.5,noise=.09)
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
km = KMeans(n_clusters=2, random_state=0)
y_km = km.fit_predict(X)
```

```python
plt.scatter(X[y_km==0,0],X[y_km==0,1], c='lightblue', label = "Cluster1")
plt.scatter(X[y_km==1,0],X[y_km==1,1], c='red', label = "Cluster2")
plt.title("K-means")
plt.legend(loc=(0.77,1.1))
plt.show()
```
