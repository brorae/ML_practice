```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import DBSCAN
from sklearn.datasets import make_circles
```

```python
X,y = make_circles(n_samples=500,factor=.5,noise=.05)
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
db = DBSCAN(eps=0.2, min_samples=5, metric='euclidean')
y_db = db.fit_predict(X)
```

```python
plt.scatter(X[y_db==0,0],X[y_db==0,1], c='lightblue', label = "Cluster1")
plt.scatter(X[y_db==1,0],X[y_db==1,1], c='red', label = "Cluster2")
plt.title("DBSCAN")
plt.legend(loc=(0.77,1.1))
plt.show()
```
